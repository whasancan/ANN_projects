```python
import pandas as pd
import tensorflow as tf
import numpy as np

from sklearn.preprocessing import LabelEncoder

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler

from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
data = pd.read_csv("C:/Users/whasa/Desktop/Churn_Modelling.csv")
data.drop(["RowNumber", "CustomerId", "Surname"], axis=1, inplace=True)


data.head()
```

```python
x = data.iloc[:,0:-1].values
y= data.iloc[:,-1].values
```

```python
# cinsiyeti 0-1 yapmak için LabelEncoder kullandık

le = LabelEncoder()
x[:,2]= le.fit_transform(x[:,2])
```

```python
# ülkeleri sayıya çevirmek için OneHotEncoder kullanıcaz label kullanırsak ezber yapabilir
# ülke sütunun kaldırır onun yerien keimin memlketi ise ona 1 yazar

ct = ColumnTransformer(transformers=[("encoder", OneHotEncoder(), [1])], remainder="passthrough")
x=np.array(ct.fit_transform(x))
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
# çok büyük sayılar olduğu için grafikde düzgün gözükmesi için StandardScaler yapacağız

sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
# yapay sinir ağın varlığını bildiriyoruz

ann = tf.keras.models.Sequential()
```

```python
# kerasın altında katman oluşturuyoruz
# relu methodunu kullanıyoruz

ann.add(tf.keras.layers.Dense(units=6, activation="relu"))
```

```python
# çıkkış katamı

ann.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
# neural networku çalıştırıyoruz

ann.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
# veri setini fit edelim

ann.fit(xtrain, ytrain, epochs=100)
```

```python
ypred= ann.predict(xtest)
```

```python
ypred = (ypred>0.5)
```

```python
print("%",accuracy_score(ytest,ypred)*100)
```
